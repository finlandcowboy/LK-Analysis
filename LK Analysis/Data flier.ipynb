```python
import os
cwd = os.getcwd()
if 'projects.xls' in os.listdir(cwd):
    os.remove(cwd + '/' + 'projects.xls')
```

```python
import xlwt
font = xlwt.easyxf('font: height 120, name Calibri,colour_index black, bold off,\
    italic off; align: wrap on, vert top, horiz left;\
    pattern: pattern solid, fore_colour red;') 
work_data = xlwt.Workbook('utf-8')
sheet = work_data.add_sheet('всё', cell_overwrite_ok = True)
```

```python
import datetime
time = datetime.datetime.today().strftime('%d.%m')
```

```python
#подгрузка
dates_list = []
path = '/Users/qeepnet/Desktop/urls'
for elem in os.listdir(path):
    try: 
        datetime.datetime.strptime(elem or "", "%d.%m")
        dates_list.append(elem)
    except:
        pass
for i in range(len(dates_list)):
    for j in range(len(dates_list)):
        if dates_list[i] < dates_list[j]:
            dates_list[i], dates_list[j] = dates_list[j], dates_list[i]
```

```python
counter = 0
for date in dates_list:
    count = 1
    sheet.write(0,counter,str(date))
    projects = os.listdir(path + '/' + str(date))
    for project in projects:
        if not project.endswith('копия') and not project.endswith('2') and not project.startswith('.DS_S'):
            if project.endswith('.txt'):
                project = project[:-4]
            sheet.write(count,counter,str(project))
            count = count + 1
    counter += 1
```
